```python
url='https://raw.githubusercontent.com/terranigmark/curso-analisis-exploratorio-datos-platzi/main/train_titanic.csv'
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df_titanic = pd.read_csv(url, error_bad_lines=False)
```

```python
df_titanic
```

```python
df_titanic.info()
```

```python
df_titanic["Survived2"] = df_titanic["Survived"].map({0:"not_survived", 1:"survived"})
```

```python
df_titanic["Pclass2"] = df_titanic["Pclass"].map({"C":1, "S":2,"Q":3})
```

```python
sex_surv= pd.crosstab(df_titanic["Sex"], df_titanic["Survived2"])
```

```python
sex_surv
```

```python
sns.barplot(data= df_titanic,x= "Pclass", y= "Survived", hue="Sex")
plt.legend()
plt.show()
```

```python
df_titanic["Sex2"] = df_titanic["Sex"].map({"male":0, "female":1})
```

```python
df_titanic_short= df_titanic.groupby(["Age","Pclass","Survived","Sex"])
```

```python
sns.barplot(data= df_titanic,x= "Pclass", y= "Survived")
plt.show()
```
